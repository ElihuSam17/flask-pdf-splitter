## Código base
Cambie SOLO los nombres que estén en "" según su preferencia.

```python
import os
import pandas as pd
from PyPDF2 import PdfReader, PdfWriter

# Archivos de entrada
pdf_file = "/Constancias de la UAEH.pdf"  # Asegúrate de que esta ruta sea correcta
excel_file = "/participes.xlsx"     # Asegúrate de que esta ruta sea correcta
#########################################################
output_folder = "/content/constancias UAEH AA"#Cambiar cada que se cambie de curso
#########################################################
# Crear carpeta de salida si no existe
os.makedirs(output_folder, exist_ok=True)

# Cargar nombres desde Excel
df = pd.read_excel(excel_file)
nombres = df["Nombres"].tolist()  # Asegúrate de que la columna se llama "Nombres"

# Leer el PDF
pdf_reader = PdfReader(pdf_file)
num_pages = len(pdf_reader.pages)

# Verificar que hay suficientes nombres para cada página del PDF
if len(nombres) != num_pages:
    raise ValueError(f"Error: El número de nombres ({len(nombres)}) no coincide con el número de páginas ({num_pages}) en el PDF.")

# Separar cada página y guardarla con el nombre del participante
for i, nombre in enumerate(nombres):
    pdf_writer = PdfWriter()
    pdf_writer.add_page(pdf_reader.pages[i])
############################################################
    pdf_filename = os.path.join(output_folder, f"Constancias UAEH AA-{nombre}.pdf")# Cambiar cada que se ejecute un nuevo curso
#############################################################
    with open(pdf_filename, "wb") as pdf_output:
        pdf_writer.write(pdf_output)

    print(f"Generado: {pdf_filename}")

print("✅ ¡Todos los PDFs han sido generados correctamente!")
```

# Código para descargar carpeta .zip
Ejecute este código hasta que desee descargar la carpeta que generó.

```python
import shutil

# Nombre del archivo ZIP
#############################################################
zip_filename = "/content/Constancias AA.zip"#Cambiar cada que se cambie de curso
#############################################################
# Comprimir la carpeta en un ZIP
shutil.make_archive(zip_filename.replace(".zip", ""), 'zip', output_folder)

# Descargar el archivo ZIP
from google.colab import files
files.download(zip_filename)
```
